***W9A1***

Viết hàm merge hai danh sách đã sắp xếp mà không dùng bất cứ thuật toán sắp xếp nào

```python
def merge(l1,l2):
  kq=[]
  i=0
  j=0
  while i<len(l1) and j<len(l2):
    if l1[i]<l2[j]:
      kq.append(l1[i])
      i+=1
    else:
      kq.append(l2[j])
      j+=1
  if i<len(l1):
    kq.extend(l1[i:])
  if j<len(l2):
    kq.extend(l2[j:])
  return kq
l1=list(map(int,input().split()))
l2=list(map(int,input().split()))
a=merge(l1,l2)
for i in a:
  print(i,end=' ')
```

***W9A3***

Cho một số nguyên dương, hãy thực hiện hoán đổi đúng 2 chữ số để được một số nguyên nhỏ nhất có thể


```python
def swap_small(n):
  s=list(str(n))
  l=len(s)
  for i in range (l-1):
    x=s[i+1:]
    m=min(x)
    if m<s[i]:
      for j in range(l-1,i,-1):
        if s[j]==m:
          s[i],s[j]=s[j],s[i]
          return int(''.join(s))
  return n
n=int(input())
print(swap_small(n))
```

Cho xâu a gồm các chữ cái thường và in hoa và xâu b chỉ gồm
các chữ cái in hoa. Ta có thể biến đổi xâu a theo 2 bước như sau:
- Chuyển một số chữ cái thường trong xâu a thành chữ cái in hoa
(có thể không biến đổi chữ cái nào)
- Xóa hết tất cả các chữ cái thường trong xâu a.
Cho 2 xâu a,b, xác đính xem có cách biến đổi từ xâu a sang xâu
b không. Ghi ra "YES" nếu có và "NO" nếu ngược lại.


```python
def biendoi(a,b):
  n=len(a)
  m=len(b)
  dp=[[False]*(m+1) for _ in range(n+1)]
  dp[0][0]=True
  for i in range (1,n+1):
    if a[i-1].islower():
      dp[i][0]=dp[i-1][0]
    else:
      dp[i][0]=False
  for i in range (1,n+1):
    for j in range(1,m+1):
      c1=a[i-1]
      c2=b[j-1]
      if c1==c2:
        dp[i][j]=dp[i-1][j-1]
      elif c1.islower():
        dl=dp[i-1][j]
        op=False
        if c1.upper()==c2:
          op=dp[i-1][j-1]
        dp[i][j]=dl or op
      else:
        dp[i][j]=False
  return "YES" if dp[n][m] else "NO"
q=int(input())
for _ in range(q):
  a=input()
  b=input()
  print(biendoi(a,b))
  q-=q
```

***W9A5***

Cho hai mảng số nguyên list1 và list2, viết hàm kiểm tra xem list1
có thuộc list2 hay không (có phải là dãy con liên tiếp của list2 hay
không)?

```python
def lientiep(l1,l2):
  if len(l1)==0:
    return True
  if len(l1)>len(l2):
    return False
  for i in range (len(l2)-len(l1)+1):
    if l2[i:i+len(l1)]==l1:
      return True
  return False
l1=list(map(int,input().split()))
l2=list(map(int,input().split()))
if (lientiep(l1,l2)):
  print("YES")
else:
  print("NO")
```

[Hoán vị tiếp theo]
Cho một dãy số gồm n chữ số là một hoán vị của dãy 1,2,...,n.
Viết chương trình in ra hoán vị ở vị trí tiếp theo (danh sách các
hoán vị được sắp xếp theo thứ tự từ điển) của hoán vị đã cho.
Đầu vào từ bàn phím gồm 2 dòng:
- Dòng đầu tiên chứa số nguyên n
- Dòng tiếp theo chứa n số nguyên là một hoán vị của dãy 1,2,...,n
(n≤105)
Đầu ra: In ra màn hình hoán vị tiếp theo của hoán vị đã cho.

```python
def hoanvi(a):
  n=len(a)
  #Tìm vị trí i sao cho a[i]<a[i+1]
  #Duyệt từ phần tử áp chót về đầu
  i=n-2
  while i>=0 and a[i]>=a[i+1]:
    i-=1
  #Nếu tìm thấy i (không phải hoán vị cuối cùng)
  #VD: 12345 -> 12354
  if i>=0:
    #Tìm j: a[j]>a[i]
    j=n-1
    while a[j]<=a[i]:
      j-=1
    #Đổi chỗ a[i],a[j]
    a[i],a[j]=a[j],a[i]
  #Lật ngược đoạn từ i+1 đến cuối
  l=i+1
  r=n-1
  while l<r:
    a[l],a[r]=a[r],a[l]
    l+=1
    r-=1
  return a
k=int(input())
a=list(map(int,input().split()))
b=hoanvi(a)
for i in b:
  print(i,end=' ')
```

[CanPlaceFlowers]
Giả sử bạn là nhân viên của một công ty cây xanh, được giao cho
chăm sóc một luống đất. Luống đất đó được chia thành nhiều ô
đất nhỏ có kích cỡ bằng nhau. Trước khi được giao cho bạn
chăm sóc, trên luống hoa đã được trồng sẵn một số cây xanh tán
rộng. Một ngày đầu xuân, công ty gửi đến cho bạn một số cây
hoa và yêu cầu bạn trồng vào những vị trí còn trống trong luống
hoa. Đây là những cây hoa hướng sáng và chúng không thể sống
được dưới bóng râm của cây khác, đặc biệt là cây tán rộng. Vì
vậy, bạn không thể trồng chúng tại ô bên cạnh những ô đã được
trồng cây xanh tán rộng. Và nếu trồng những cây hoa bên cạnh
nhau, chúng sẽ tranh giành phân bón và nước và sau cùng, cả
hai cây sẽ cùng chết.
Viết hàm canPlaceFlowers(flowerbed, k) kiểm tra xem có thể
trồng hết số hoa mới được cấp vào trong luống đất của bạn hay
không.
Hàm nhận đầu vào là một danh sách (list) flowerbed, đánh dấu
những vị trí trong luống đất đã được trồng cây xanh trước đó (giá
trị 1 ứng với vị trí cây đã được trồng và 0 ứng với vị trí đất trống)
và số lượng hoa được phát thêm k (int). Hàm trả về True nếu có
thể trồng hết toàn bộ k cây hoa vào luống đất và False trong
trường hợp ngược lại.


```python
def canPlaceFlowers(flowerbed, k):
  #Kiểm tra xem có thể trồng k bông hoa vào flowerbed mà không vi phạm quy tắc
  n=len(flowerbed)
  for i in range(n):
    #Nếu đã trồng đủ số hoa thì trả về True ngay lập tức
    if k==0:
      return True
    #Kiểm tra ba điều kiện để trồng được hoa tại i:
    #1. Ô hiện tại phải trống (flowerbed[i]==0)
    #2. Ô bên trái trống hoặc i là ô đầu tiên (i==0)
    #3. Ô bên phải trống hoặc i là ô cuối cùng (i==n-1)
    if flowerbed[i]==0:
      left_empty=(i==0) or (flowerbed[i-1]==0)
      right_empty=(i==n-1) or (flowerbed[i+1]==0)
      if left_empty and right_empty:
        flowerbed[i]=1
        k-=1
  return k<=0
flowerbed=list(map(int,input().split()))
k=int(input())
print(canPlaceFlowers(flowerbed,k))
```

[Phát hiện đường đi lặp lại của Rô-bốt]
Ban đầu, một con rô-bốt ở vị trí (0, 0) trên một lưới 2D.
Cho một chuỗi `moves` chứa các thao tác di chuyển của rô-bốt.
Các bước di chuyển được biểu diễn bởi các ký tự:
- 'R' (Phải)
- 'L' (Trái)
- 'U' (Lên)
- 'D' (Xuống)
Viết hàm `def is_path_crossing(moves)`:
Hàm nhận đầu vào là một chuỗi `moves`.
Hàm trả về `True` nếu trong quá trình di chuyển, rô-bốt đi vào một mà nó đã từng đi qua trước đó.
Hàm trả về `False` nếu rô-bốt không bao giờ đi lặp lại (luôn đi vào
ô mới).
Lưu ý: Vị trí ban đầu (0, 0) được tính là vị trí đầu tiên đã đi qua.




```python
def is_path_crossing(moves):
  x,y=0,0
  visited={(0,0)}
  for move in moves:
    if move=='U':
      y+=1
    elif move=='D':
      y-=1
    elif move=='R':
      x+=1
    elif move=='L':
      x-=1
  current=(x,y)
  if current in visited:
    return True
  visited.add(current)
  return False
path=input()
print(is_path_crossing(path))
```

Viết hàm `def Median(nums):`
Hàm nhận đầu vào là một danh sách (list) `nums` được đảm bảo
có đúng 5 số nguyên.
Yêu cầu:
Tìm và trả về số trung vị (số lớn thứ ba) của 5 số này.
Ràng buộc:
1. Chỉ được sử dụng tối đa 6 phép so sánh (ví dụ: <, >, <=, >=).
2. KHÔNG ĐƯỢC DÙNG bất kỳ hàm sắp xếp nào (như
`sorted()`, `list.sort()`).
Chú ý: Bạn chỉ cần viết hàm `Median` theo yêu cầu.
Phần đọc input và gọi hàm (hàm main) đã có sẵn.


```python
def Median(nums):
  if nums[0]>nums[1]:
    nums[0],nums[1]=nums[1],nums[0]
  if nums[2]>nums[3]:
    nums[2],nums[3]=nums[3],nums[2]
  if nums[0]>nums[2]:
    nums[0],nums[2]=nums[2],nums[0]
    nums[1],nums[3]=nums[3],nums[1]
  if nums[1]>nums[4]:
    nums[1],nums[4]=nums[4],nums[1]
  if nums[1]>nums[2]:
    nums[1],nums[2]=nums[2],nums[1]
    if nums[2]<nums[3]:
      return nums[2]
    else:
      return nums[3]
  else:
    if nums[2]<nums[4]:
      return nums[2]
    else:
      return nums[4]
nums=list(map(int,input().split()))
print(Median(nums))
```

[Tìm bán kính lò sưởi nhỏ nhất]
Mùa đông đang đến! Bạn được cung cấp vị trí của các ngôi nhà
và các lò sưởi trên một đường thẳng.
Nhiệm vụ của bạn là tìm bán kính r nhỏ nhất cho các lò sưởi sao
cho mọi ngôi nhà đều nằm trong phạm vi sưởi ấm của ít nhất một
lò sưởi.
Tất cả các lò sưởi đều có cùng một bán kính r.
Một ngôi nhà tại vị trí h được sưởi ấm bởi một lò sưởi tại vị trí t
nếu |h - t| <= r.
Viết hàm `def findRadius(houses, heaters):`
Hàm nhận đầu vào là hai danh sách (list) các số nguyên:
- `houses`: chứa vị trí của các ngôi nhà.
- `heaters`: chứa vị trí của các lò sưởi.
Hàm trả về một số nguyên là bán kính r nhỏ nhất cần thiết.
Lưu ý:
- Số lượng nhà và lò sưởi là một số không âm và không vượt quá 25000.
- Vị trí của các ngôi nhà và lò sưởi là một số không âm và không
vượt quá 10^9.

```python
def findRadius(houses,heaters):
  houses.sort()
  heaters.sort()
  radius=0
  i=0
  for house in houses:
    while i<len(heaters) -1 and abs(heaters[i+1]-house)<=abs(heaters[i]-house):
      i+=1
    curr=abs(heaters[i]-house)
    radius=max(radius,curr)
  return radius
houses=list(map(int,input().split()))
heaters=list(map(int,input().split()))
print(findRadius(houses,heaters))
```
