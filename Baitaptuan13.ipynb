```python
#W11A1
import os
def reverse(path:str) ->int:
    #Mở file để đọc
    with open(path,'r',encoding='utf-8') as f:
        a=f.read()
        num=a.strip()
        reve=num[::-1]
        return int(reve)
if __name__=="__main__":
    filename="test_number.txt"
    with open(filename,'w',encoding='utf-8') as f:
        f.write("1503")
    kq=reverse(filename)
    print(f"Số đảo ngược: {kq}")
```

```python
#W11A2
import os
def canWinNim(path:str) ->bool:
    #Mở file để đọc
    with open(path,'r',encoding='utf-8') as f:
        n=int(f.read().strip())
    return n%4!=0
if __name__=="__main__":
    filename="Win_Nim.txt"
    with open(filename,'w',encoding='utf-8') as f:
        f.write("4")
    if canWinNim(filename):
        print("WIN")
    else:
        print("LOSE")
```

```python
#W11A4
def areAnagrams(path1:str,path2:str):
    def get_clean(text):
        return sorted([char.lower() for char in text if char.isalpha()])
    with open(path1.strip(),'r',encoding='utf-8') as f1:
        con1=f1.read()
    with open(path2.strip(),'r',encoding='utf-8') as f2:
        con2=f2.read()
    if get_clean(con1)==get_clean(con2):
        return True
    else:
        return False
if __name__=="__main__":
    filename1="Hogwarts1.txt"
    filename2="Hogwarts2.txt"
    with open(filename1,'w',encoding='utf-8')as f:
        s1=input().strip()
        f.write(s1)
    with open(filename2,'w',encoding='utf-8')as f:
        s2=input().strip()
        f.write(s2)
    result=areAnagrams(filename1,filename2)
    print(result)
```

```python
#W11A8
def tu_fibonacci():
    try:
        n=int(input())
        if n>20 or n<0:
            print("Nhập lại")
            return
        fib=[""]*(n+1)
        if n>=1:
            fib[1]="a"
        if n>=2:
            fib[2]="b"
        for i in range(3,n+1):
            fib[i]=fib[i-1]+fib[i-2]
        with open("output.txt",'w',encoding='utf-8') as f:
            for i in range(1,n+1):
                line=f"f({i})= {fib[i]}"
                print(line)
                f.write(line+"\n")
    except ValueError:
        print("Vui lòng nhập số nguyên hợp lệ")
tu_fibonacci()
```

```python
#W11A10
def getMoney(path:str)->list[int]:
    money_list=[]
    with open(path,'r',encoding='utf-8') as f:
        con=f.read()
        items=con.split()
        money_list=[int(item) for item in items]
    return money_list
def robbery(money:list[int]) -> int:
    if not money:
        return 0
    if len(money)==1:
        return money[0]
    dp=[0]*len(money)
    dp[0]=money[0]
    dp[1]=max(money[0],money[1])
    for i in range (2,len(money)):
        dp[i]=max(dp[i-1],dp[i-2]+money[i])
    return dp[-1]
with open('money.txt','w') as f:
    f.write("1 2 3 1")
dstien=getMoney('money.txt')
kq=robbery(dstien)
print(kq)
```

```python
#W11A11
def max_min(path:str):
    try:
        a=[]
        with open(path,'r',encoding='utf-8') as f:
            con=f.read()
            b=con.split()
            a=[int(i) for i in b]
        return max(a),min(a)
    except FileNotFoundError:
        return False
filename=input()
with open(filename,'w',encoding='utf-8') as f:
    s=input()
    f.write(s)
if max_min(filename)==False:
    print("Mission failed")
else:
    a,b=max_min(filename)
    print(f"max: {a}; min: {b}")
```

```python
#W11A12
def max3so(path:str)->int:
    num=[]
    with open(path,'r',encoding='utf-8') as f:
        con=f.read()
        b=con.split()
        num=[int(i) for i in b]
        num.sort()
        n=len(num)
        if n<3:
            return 0
        max1=num[n-1]*num[n-2]*num[n-3]
        max2=num[0]*num[1]*num[n-1]
    return max(max1,max2)
filename='data.txt'
with open(filename,'w',encoding='utf-8') as f:
    s=input()
    f.write(s)
print(max3so(filename))
```

```python
#W11A13
def moveZeroes(path:str)->int:
    num=[]
    with open(path,'r',encoding='utf-8') as f:
        con=f.read()
        b=con.split()
        dem=0
        for i in b:
            if int(i)==0:
                dem+=1
            else:
                num.append(int(i))
        kq=num+[0]*dem
    return kq
filename='data.txt'
with open(filename,'w',encoding='utf-8') as f:
    s=input()
    f.write(s)
zero=moveZeroes(filename)
for i in zero:
    print(i,end=" ")
```

```python
#W11A14
import csv
def average_time(path:str):
    total_time=0
    count=0
    with open(path,'r',encoding='utf-8') as f:
        reader=csv.reader(f)
        header=next(reader)
        for row in reader:
            if len(row)>=2:
                time=float(row[1])
                total_time+=time
                count+=1
        if count==0:
            return 0.0
        avg=total_time/count
        return avg
filename='average_time.csv'
content="""Lap,Time
1,143.23
2,134.12"""
with open(filename,'w',encoding='utf-8') as f:
    f.write(content)
kq=average_time(filename)
print(f"{kq:.2f}")
```

```python
#W11A16
import math

def chuvi(path: str):
    with open(path,'r',encoding='utf-8') as f:
        lines=[line.strip() for line in f if line.strip()]
        i=0
        while i<len(lines):
            line=lines[i]
            if line.startswith("TYPE:"):
                type=line.split(":")[1].strip()
                if i+1<len(lines):
                    dim_line=lines[i+1]
                    chuvi=0
                    if type == "SQUARE":
                        side = float(dim_line.split(":")[1])
                        chuvi = 4 * side
                    elif type == "RECTANGLE":
                        parts = dim_line.split()
                        w = float(parts[1])
                        h = float(parts[3])
                        chuvi = 2 * (w + h)
                    elif type == "CIRCLE":
                        if "RADIUS" in dim_line:
                            r = float(dim_line.split(":")[1])
                            chuvi= 2 * math.pi * r
                    print(f"{chuvi:.2f}")
                    i += 1
            i += 1
content = """
TYPE: SQUARE
SIDE: 5.3

TYPE: RECTANGLE
WIDTH: 5.0 HEIGHT: 16.0

TYPE: CIRCLE
RADIUS: 2.15"""
with open('shapes.txt', 'w', encoding='utf-8') as f:
    f.write(content)
chuvi('shapes.txt')
```

```python
#W11A18
import math
def estimatedTime(path:str):
    with open(path,'r',encoding='utf-8') as f:
        lines=[line.strip() for line in f if line.strip()]
        if len(lines)<3:
            return 0.0
        parts_src=lines[0].split()
        x1=float(parts_src[2])
        y1=float(parts_src[4])
        parts_des=lines[1].split()
        x2=float(parts_des[2])
        y2=float(parts_des[4])
        parts_v=lines[2].split()
        v=float(parts_v[1])
        dis=math.sqrt((x2-x1)**2 +(y2-y1)**2)
        if v==0:
            return 0.0
        time=dis/v
        print(f"{time:.2f}")
filename='data.txt'
content="""
SRC X: 0 Y: 0
DES X: 10000 Y: 10000
VELOCITY: 6 KM/S"""
with open(filename,'w',encoding='utf-8') as f:
    f.write(content)
estimatedTime(filename)
```

```python
#W11A19
import struct
import os

def read_binary_doubles(path:str):
    with open(path, 'rb') as f:
        while True:
            chunk = f.read(8)
            if len(chunk) < 8:
                break
            value = struct.unpack('d', chunk)[0]
            print(f"{value:.5f}")
def create_test_file(filename):
    data = [3.1415926535, 2.718281828, 100.5, -12.3456789]
    with open(filename, 'wb') as f:
        for num in data:
            f.write(struct.pack('d', num))
if __name__ == "__main__":
    test_filename = 'numbers.bin'
    create_test_file(test_filename)
    read_binary_doubles(test_filename)
```

```python
#W11A20
import math
def imageSmoother(path:str):
    with open(path,'r',encoding='utf-8') as f:
        line1=[line.strip() for line in f if line.strip()]
        if not line1:
            return
        M=len(line1)
        img=[]
        for line in line1:
            row=[int(x) for x in line.split()]
            img.append(row)
    res=[[0]*M for _ in range(M)]
    for r in range(M):
        for c in range(M):
            total=0
            count=0
            for i in range(r-1,r+2):
                for j in range(c-1,c+2):
                    if 0<=i<M and 0<=j<M:
                        total+=img[i][j]
                        count+=1
            res[r][c]=math.floor(total/count)
    for row in res:
        print(row)
filename='image_data.txt'
content="""
130 160 150 170
160 140 180 160
170 190 120 150
210 180 160 120"""
with open(filename,'w',encoding='utf-8') as f:
    f.write(content)
imageSmoother(filename)
```
