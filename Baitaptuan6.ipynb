***W6A4***

Nhập một dòng gồm các cặp key:value cách nhau
bởi dấu cách. Tạo dictionary ánh xạ mỗi key -> list
các value theo thứ tự xuất hiện. Nếu key lặp lại,
thêm vào list của key đó

```python
cap=input().split()
a={}
for x in cap:
  key,value=x.split(':')
  value=int(value)
  if key not in a:
    a[key]=[value]
  else:
    a[key].append(value)
print(a)
```

***W6A9***

Viết chương trình nhập vào hai dictionary. Sau đó gộp lại
thành một dictionary mới trong đó nếu một key xuất hiện
trong cả hai dict, giá trị sẽ được cộng lại, nếu chỉ có trong
một dict thì giữ nguyên. In ra dictionary mới theo thứ tự
từ điển tăng dần của key.

```python
import ast
a=ast.literal_eval(input())
b=ast.literal_eval(input())
d={}
d=a.copy()
for k,v in b.items():
  if k in d:
    d[k]+=v
  else:
    d[k]=v
d=dict(sorted(d.items()))
print(d)
```

***W6A10***

Viết chương trình nhập vào một danh sách các số nguyên,
và số nguyên dương k. In ra list gồm các cặp số khác
nhau có tổng bằng k. Mỗi cặp là một tuple, hai số trong
tuple có thể giống nhau, thứ tự xuất hiện trong danh sách
của số thứ nhất nhỏ hơn thứ tự của số thứ hai. Các tuple
trong list được sắp xếp theo thứ tự tăng dần của phần tử
đầu tiên trong tuple

```python
num=list(map(int,input().split()))
k=int(input())
res=[]
for i in range (len(num)):
  for j in range(i+1,len(num)):
    if num[i]+num[j]==k:
      res.append((num[i],num[j]))
res.sort()
print(res)
```

***W6A12***

[Tìm phần tử xuất hiện nhiều nhất]
Nhập vào một list các số nguyên. Hãy tìm số xuất
hiện nhiều lần nhất trong list. Nếu có nhiều số cùng
có số lần xuất hiện nhiều nhất, hãy in ra số nhỏ nhất
trong số đó

```python
a=map(int,input().split())
count={}
for i in a:
  if i not in count:
    count[i]=1
  else:
    count[i]+=1
result=0
tam=0
for key,value in count.items():
  if value>tam:
    result=key
    tam=value
  elif value == tam:
    result=min(result,key)
print(result)
```

số nguyên, và một số nguyên k ở dòng tiếp theo.

***W6A17***

Nhập vào mảng vuông 2 chiều n hàng, n cột, in ra
đường chéo chính và đường chéo phụ của mảng


```python
n=int(input())
a=[]
for i in range(n):
  row=[]
  x=map(int,input().split())
  for t in x:
    row.append(t)
  a.append(row)
print("Đường chéo chính")
for i in range(n):
  print(a[i][i],end=" ")
print()
print("Đường chéo phụ")
for i in range(n):
  print(a[i][n-1-i],end=" ")
```

W6A18

Nhậ̣p vào mảng n hàng, m cột số nguyên và số
nguyên k (n, m, k nhập trên cùng một dòng, cách
nhau dấu cách). Tính và in ra tổng các phần tử trên
cột k (hàng k)

```python
n,m=map(int,input().split())

a=[]
for i in range(n):
  row=[]
  x=map(int,input().split())
  for t in x:
    row.append(t)
  a.append(row)
k=int(input())
sum_row=0
sum_col=0
for i in range(n):
  sum_col+=a[i][k-1]
for j in range(m):
  sum_row+=a[k-1][j]
print(sum_row,sum_col)
```
