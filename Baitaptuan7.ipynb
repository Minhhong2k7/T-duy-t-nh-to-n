***W7A1:***

Viết hàm tìm kiếm nhị phân
trên một danh sách đã
được sắp xếp tăng dần.
Trả về chỉ số của phần tử
hoặc -1 nếu không tìm
thấy. mẫu hàm: def
binary_search(arr, left,
right, target):

```python
def binary_search(arr,left,right,target):
  if right<left: return -1
  else:
    mid=(left+right)//2
    if arr[mid]==target: return mid
    elif arr[mid]>target: return binary_search(arr,left,mid-1,target)
    else: return binary_search(arr,mid+1,right,target)
arr=list(map(int,input().split()))
target=int(input())
print(binary_search(arr,0,len(arr)-1,target))
```

***W7A3:***

Viết hàm thực hiện thuật
toán bubble sort, Selection
Sort để sắp xếp danh sách
theo thứ tự tăng dần. Sắp
xếp nổi bọt (Bubble Sort); Sắp xếp chọn (Selection
Sort)


```python
def bubblesort(arr):
  n=len(arr)
  for i in range(n):
    swap=False
    for j in range(0,n-i-1):
      if arr[j]>arr[j+1]:
        arr[j],arr[j+1]=arr[j+1],arr[j]
        swap=True
    if not swap:
      break
a=list(map(int,input().split()))
bubblesort(a)
print(*a)
```

```python
def selectionsort(arr):
  n=len(arr)
  for i in range(n-1):
    min=i
    for j in range(i+1,n):
      if arr[j]<arr[min]:
        min=j
    arr[i],arr[min]=arr[min],arr[i]
a=list(map(int,input().split()))
selectionsort(a)
print(*a)
```

***W7A4:***

Viết hàm tìm phần tử có số
lần xuất hiện nhiều nhất
trong danh sách. Nếu có
nhiều phần tử cùng số lần,
trả về phần tử xuất hiện
trước.


```python
def frequent(a):
  freq={}
  max_count=0
  result=a[0]
  for x in a:
    freq[x]=freq.get(x,0)+1
    if freq[x]>max_count:
      max_count=freq[x]
      result=x
  return result,max_count
a=list(map(int,input().split()))
value,count=frequent(a)
print(value,"xuat hien nhieu nhat, som nhat,",count,"lan")
```

***W7A7:***

Cho một danh sách các số
nguyên nums, hãy tìm độ
dài của dãy con tăng
nghiêm ngặt dài nhất
(không nhất thiết phải liền kề).


```python
def LIS(nums):
  if not nums:
    return 0
  n=len(nums)
  dp=[1]*n
  for i in range(n):
    for j in range(i):
      if nums[j]<nums[i]:
        dp[i]=max(dp[i],dp[j]+1)
  return max(dp)
nums=eval(input())
print(LIS(nums))
```

***W7A8:***

Bạn được cung cấp: một
danh sách các khoảng thời
gian interval = [[l1, r1], [l2,
r2], ...] và một danh sách
các truy vấn queries = [q1,
q2, ...]
Với mỗi truy vấn q, hãy tìm
độ dài nhỏ nhất của
khoảng thời gian [li, ri] sao
cho li ≤ q ≤ ri.
Nếu không tồn tại khoảng
thời gian nào như vậy, hãy
trả về -1 cho truy vấn đó.
solve in < O(m.n), hint:
binary search in the
queries list


```python
#Giải thích: Cho khoảng thời gian(intervals) và truy vấn (queries).
#Với mỗi q, tìm độ dài nhỏ nhất trong khoảng [l,r]: 1<=q<=r. Không có trả về -1

import heapq #priority queue
import ast #chuyển chuỗi thành dạng
def min_interval(intervals, queries):
    intervals.sort() #sắp xếp
    qs = sorted((val, i) for i, val in enumerate(queries)) #gắn vị trí theo số
    ans = [-1] * len(queries)
    h = []
    i = 0
    for k, idx in qs:
        while i < len(intervals) and intervals[i][0] <= k:
            l, r = intervals[i]
            heapq.heappush(h, (r - l + 1, r))
            i += 1

        while h and h[0][1] < k:
            heapq.heappop(h)

        if h:
            ans[idx] = h[0][0]

    return ans
intervals=ast.literal_eval(input())
queries=ast.literal_eval(input())
print(min_interval(intervals,queries))
```

***W7A9:***

Cho một danh sách các số
nums, với mỗi phần tử
nums[i], hãy đếm xem có
bao nhiêu phần tử bên phải
nhỏ hơn nums[i].
Trả về danh sách các số
đếm.
Efficient solution: using
heap or binary search tree

```python

import heapq
def count_smaller(nums):
  heap=[]
  res=[]
  for x in reversed(nums):
    cnt=0
    temp=[]
    #Lấy các số <x
    while heap and heap[0]<x:
      temp.append(heapq.heappop(heap))
      cnt+=1
    #Đưa vào heap
    for v in temp:
      heapq.heappush(heap,v)
    #Thêm x vào heap
    heapq.heappush(heap,x)
    res.append(cnt)
  return res[::-1]
num=eval(input())
print(count_smaller(num))
```

Gợi ý:
Duy trì tổng tiền tố (prefix

```python
import bisect #Tìm vị trí chèn bằng tìm kiếm nhị phân
def max_mod(nums,m):
  pre=0
  ans=0
  sorted_pre=[]
  for x in nums:
    pre=(pre+x)%m #Tổng từ đầu đến vị trí hiện tại (Dùng tổng tiền tố mod m)
    ans=max(ans,pre)
    id=bisect.bisect_right(sorted_pre,pre)
    #Tìm prefix nhỏ nhất nhưng lớn hơn prefix hiện tại. (sorted_pe đã sắp xếp)
    #VD: [1,3,3,6] bisect_right(a,3) -> [1,3,3,|,6] -> 3 (Vị trí)
    if id<len(sorted_pre):
      ans=max(ans,(pre-sorted_pre[id]+m)%m) #Tính tổng mảng con tối ưu
      bisect.insort(sorted_pre,pre) #insort=bisect+insert
  return ans
num=eval(input())
m=int(input())
print(max_mod(num,m))
```

***W7A13:***

Dãy hài hòa là một dãy
thỏa mãn điều kiện giá trị
lớn nhất và giá trị nhỏ nhất
của dãy khác nhau đúng
bằng \( 1 \).
Viết hàm trả về độ dài của
dãy hài hòa con dài nhất có
thể có trong mảng giá trị
nguyên \( arr \) có \( n \)
phần tử
Ví dụ:
Input: 1 3 2 2 5 2 3 7
Output: 5 - Dãy hài hòa con
dài nhất là [3,2,2,2,3]

```python
def haihoa(a):
  freq={}
  for x in a:
    freq[x]=freq.get(x,0)+1
  ans=0
  for x in freq:
    if x+1 in freq:
      ans=max(ans,freq[x]+freq[x+1])
  return ans
a=list(map(int,input().split()))
print(haihoa(a))
```
