[Cylinder]

Xây dựng lớp Cylinder (hình trụ) gồm: bán kính và
chiều cao (kiểu số thực) với hai phương thức:tính
diện tích bề mặt (diện tích toàn phần) và thể tích
của hình trụ

Đầu vào: Đầu vào từ bàn phím gồm một dòng duy nhất
chứa bán kính và chiều cao của hình trụ, cách
nhau bởi một dấu cách.

Đầu ra: In ra màn hình diện tích bề mặt và thể tích của
hình trụ (sử dụng lại những hàm vừa viết, π=3.
14).
(Kết quả làm tròn đến 2 chữ số sau dấu phẩy)

```python
class Cylinder:
  def __init__(self,r,h):
    self.r=r
    self.h=h
  def dientich(self):
    return 2*3.14*self.r**2 + 2*3.14*self.r*self.h
  def thetich(self):
    return 3.14*self.r**2*self.h
r,h=map(float,input().split())
c=Cylinder(r,h)
print(f"Diện tích: {c.dientich():.2f}")
print(f"Thể tích: {c.thetich():.2f}")
```

Xây dựng lớp Date biểu diễn ngày tháng năm với

nào đó có hợp lệ hay không và phương thức tính
ngày kế tiếp của ngày tháng năm nào đó nếu nó là

Nhập vào một ngày tháng năm, (ngày tháng năm
biểu diễn dưới dạng dd/mm/yyyy). Nếu ngày
tháng năm không tồn tại, in ra dòng chữ INVALID.
Nếu là hợp lệ in ra ngày kế tiếp của nó.

Đầu vào
Đầu vào từ bàn phim chứa dữ liệu ngày, tháng,
năm dưới dạng dd/mm/yyyy .

Đầu ra
In ra màn hình ngày kế tiếp của ngày vừa nhập
dưới dạng dd/mm/yyyy
. Nếu ngày tháng không tồn tại, in ra INVALID

```python
class Date:
  def __init__(self,d,m,y):
    self.d=d
    self.m=m
    self.y=y
  def check(self):
    month=int(self.m)
    day=int(self.d)
    year=int(self.y)
    if month<1 or month>12:
      return False
    else:
      if month==2:
        if year%400==0 or (year%4==0 and year%100!=0):
          if day<1 or day>29:
            return False
        else:
          if day<1 or day>28:
            return False
      elif month in [1,3,5,7,8,10,12]:
        if day<1 or day>31:
          return False
      else:
        if day<1 or day>30:
          return False
    return True
  def nextday(self):
    month=int(self.m)
    day=int(self.d)
    year=int(self.y)
    if month==2:
      if year%400==0 or (year%4==0 and year%100!=0):
        if day==29:
          day=1
          month=3
        else:
          day+=1
      else:
        if day==28:
            day=1
            month=3
        else:
          day+=1
    elif month in [1,3,5,7,8,10,12]:
        if month==12:
          if day==31:
            day=1
            month=1
            year+=1
        else:
          if day==31:
            day=1
            month+=1
          else:
            day+=1
    elif month in [4,6,9,11]:
      if day==30:
        day=1
        month+=1
      else:
        day+=1
    print(f"{day}/{month}/{year}")
date_string=input()
day,month,year=date_string.split("/")
if Date(day,month,year).check():
  Date(day,month,year).nextday()
else:
  print("INVALID")

```

Xây dựng lớp điểm với hai thuộc tính x, y cùng
các phương thức kiểm tra xem có phải gốc tọa dộ
không, có nằm trên trục tung không, có nằm trên
trục hoành không, tính khoảng cách từ điểm đó tới
một điểm khác

Đầu vào:
Nhập vào tọa độ một điểm M bất kỳ

Đầu ra:
Thông báo M có phải gốc tọa độ không hoặc có
nằm trên trục tung hoặc trục hoành không, khoảng
cách từ M tới gốc tọa độ là bao nhiêu (lấy 2 chữ
số phần thập phân)

```python
class Point:
  def __init__(self,x,y):
    self.x=x
    self.y=y
  def check(self):
    dx=self.x
    dy=self.y
    if (dx==0 and dy==0):
      return 1
    elif dx==0:
      return 2
    elif dy==0:
      return 3
  def khoang_cach(self):
    return (self.x**2+self.y**2)**0.5
x,y=map(float,input().split())
p=Point(x,y)
if p.check()==1:
  print("M là gốc tọa độ")
elif p.check()==2:
  print("M nằm trên trục tung")
elif p.check()==3:
  print("M nằm trên trục hoành")
print(f"Khoảng cách từ M đến gốc tọa độ: {p.khoang_cach():.2f}")
```

Xây dựng lớp Calculator (máy tính bỏ túi) với các
phương thức để thực hiện các phép toán 1. add
(cộng), 2. subtract(trừ), 3. product (nhân) và 4.
divide(chia), 5. power(hàm mũ), 6. mod (lấy phần
dư) và phương thức 7. set_numbers(cập nhật 2
số mới) nếu người dùng muốn tính toán trên 2 số
khác.

Khởi tạo lớp là hai thuộc tính gồm 2 số cần tính,
Mỗi phương thức thực hiện một phép tính và trả
về kết quả tương ứng

Tạo menu và cho người dùng tính toán tới khi gõ
"yes", "YES" hoặc "Yes" cho câu hỏi "Bạn có
muốn thoát?" mới dừng.


```python
class Calculator:
  def __init__(self,a,b):
    self.a=a
    self.b=b
  def add(self):
    return self.a+self.b
  def subtract(self):
    return self.a-self.b
  def product(self):
    return self.a*self.b
  def divide(self):
    if self.b==0:
      return "Không thể chia cho 0"
    return self.a/self.b
  def mod(self):
    if self.b==0:
      return "Không thể chia cho 0"
    return self.a%self.b
  def power(self):
    return self.a**self.b
  def set_numbers(self,a,b):
    self.a=a
    self.b=b
    print(f"Đã cập nhật 2 số mới: {self.a} và {self.b}")
try:
  num1,num2=map(float,input("Nhập hai số: ").split())
  cal=Calculator(num1,num2)
  while True:
    print("\n-----------MENU----------")
    print(f"Hai số hiện tại: {cal.a} và {cal.b}")
    print("1.Add (Cộng)")
    print("2.Subtract (Trừ)")
    print("3.Product (Nhân)")
    print("4.Divide (chia)")
    print("5.Power (hàm mũ)")
    print("6.Mod (Lấy phần dư)")
    print("7.set_numbers (cập nhật 2 số mới)")
    choice=input("Chọn chức năng (1-7): ")
    if choice=='1':
      print(f"Kết quả cộng: {cal.add()}")
    elif choice=='2':
      print(f"Kết quả trừ: {cal.subtract()}")
    elif choice=='3':
      print(f"Kết quả nhân: {cal.product()}")
    elif choice=='4':
      print(f"Kết quả chia: {cal.divide()}")
    elif choice=='5':
      print(f"Kết quả hàm mũ: {cal.power()}")
    elif choice=='6':
      print(f"Kết quả mod: {cal.mod()}")
    elif choice=='7':
      n1,n2=map(float,input("Nhập 2 số mới: ").split())
      cal.set_numbers(n1,n2)
    else:
      print("Lựa chọn không hợp lệ!")
    ask=input("\n Bạn có muốn thoát không? ")
    if ask in ["YES","yes","Yes"]:
      print("Chương trình kết thúc")
      break
except ValueError:
  print("Lỗi. Vui lòng nhập số hợp lệ!")
```

***W8A5***

Xây dựng lớp ShoppingCart (giỏ hàng) để quản lý
các sản phẩm của khách. Dùng danh sách để lưu
các sản phẩm, mỗi sản phẩm là một từ điển có
name (tên) và price(giá).
Với các phương thức sau:
1. Khởi tạo với danh sách rỗng
2. Thêm vào giỏ hàng một sản phẩm mới
3. Xóa khỏi giỏ một sản phẩm nào đó (tìm theo tên
để xóa)
4. Kiểm tra xem giỏ hàng có rỗng không
5. Tính tổng tiền giỏ hàng
6. Hiển thị toàn bộ sản phẩm trong giỏ
7. Xóa toàn bộ giỏ hàng

Tạo menu cho người dùng nhập lựa chọn truy vấn
từ 2 tới 7. Chương trình dừng khi người dùng
nhập "yes" cho câu hỏi "Có muốn thoát khỏi
chương trình không?"

```python
class ShoppingCart:
    def __init__(self):
        # 1. Khởi tạo với danh sách rỗng
        self.product = []

    def add_product(self, name, price):
        # 2. Thêm vào giỏ hàng một sản phẩm mới
        item = {"name": name, "price": price}
        self.product.append(item)
        print(f"Đã thêm '{name}' với giá '{price}' vào giỏ.")

    def remove_product(self, name):
        # 3. Xóa khỏi giỏ một sản phẩm nào đó
        found = False
        for item in self.product:
            if item["name"].lower() == name.lower():
                self.product.remove(item)
                print(f"Đã xóa sản phẩm '{name}' khỏi giỏ hàng")
                found = True
                break
        if not found:
            print(f"Không tìm thấy sản phẩm '{name}' trong giỏ hàng")

    def is_empty(self):
        return len(self.product) == 0

    def calculate_total(self):
        # 5. Tính tổng tiền giỏ hàng
        total = sum(item["price"] for item in self.product)
        return total

    def display_cart(self):
        # 6. Hiển thị toàn bộ sản phẩm trong giỏ
        if self.is_empty():
            print("Giỏ hàng đang trống")
        else:
            for i, item in enumerate(self.product, 1):
                print(f"{i}. {item['name']}: {item['price']}")

    def clear_cart(self):
        # 7. Xóa toàn bộ giỏ hàng
        self.product.clear()
        print("Đã xóa toàn bộ giỏ hàng")

def main():
    cart = ShoppingCart()
    while True:
        print("\n-----------QUẢN LÝ GIỎ HÀNG-----------")
        print("2. Thêm vào giỏ hàng một sản phẩm mới")
        print("3. Xóa khỏi giỏ một sản phẩm nào đó")
        print("4. Kiểm tra xem giỏ hàng có rỗng không")
        print("5. Tính tổng tiền giỏ hàng")
        print("6. Hiển thị toàn bộ sản phẩm trong giỏ")
        print("7. Xóa toàn bộ giỏ hàng")

        choice = input("Nhập lựa chọn của bạn (2-7): ")

        if choice == '2':
            name = input("Nhập tên sản phẩm: ")
            try:
                price = float(input("Nhập giá sản phẩm: "))
                cart.add_product(name, price)
            except ValueError:
                print("Lỗi: Giá tiền phải là số.")
        elif choice == '3':
            name = input("Nhập tên sản phẩm cần xóa: ")
            cart.remove_product(name)
        elif choice == '4':
            if cart.is_empty():
                print("Giỏ hàng đang rỗng")
            else:
                print("Giỏ hàng đang có sản phẩm")
        elif choice == '5':
            print(f"Tổng tiền của giỏ hàng: {cart.calculate_total()}")
        elif choice == '6':
            cart.display_cart()
        elif choice == '7':
            cart.clear_cart()
        else:
            print("Lựa chọn không hợp lệ. Vui lòng nhập từ (2-7)")
        ask = input("\nCó muốn thoát khỏi chương trình không? (yes/no): ")
        if ask.lower() == 'yes':
            print("Chương trình kết thúc")
            break
if __name__ == "__main__":
    main()
```

Xây dựng lớp Rectangle(w, h) với các phương
thức:

area() trả về diện tích.

perimeter() trả về chu vi.

scale(k) phóng to hình chữ nhật theo hệ số k
(nhân cả w và h), không trả về gì.

Input:
w h k (số thực dương)

Output:
area perimeter sau khi phóng to (làm tròn 2 chữ
số thập phân).


```python
class Rectangle:
  def __init__(self,w,h):
    self.w=w
    self.h=h
  def area(self):
    return self.h*self.w
  def perimeter(self):
    return 2*(self.h+self.w)
  def scale(self,k):
    self.w=self.w*k
    self.h=self.h*k
w,h,k=map(float,input().split())
rec=Rectangle(w,h)
rec.scale(k)
print(f"{rec.area():.2f} {rec.perimeter():.2f}")
```

***W8A7***

Yêu cầu

Lớp Fraction(num, den) (tử, mẫu nguyên, mẫu ≠
0) với:

Rút gọn phân số trong __init__ (quy ước mẫu >
0).

Các phương thức add/sub/mul/div trả về phân số
đã rút gọn.

__str__() in dạng a/b.

Input

a b op c d (tương ứng a/b (op) c/d, op thuộc {+,-,*,
/})

Output

Phân số kết quả đã rút gọn


```python
import math
class Fraction:
  def __init__(self,num,den):
    if den==0:
      raise ValueError("Mẫu không được bằng 0")
    if den<0:
      num=-num
      den=-den
    c=math.gcd(num,den)
    self.num=num//c
    self.den=den//c
  def add(self,other):
    #a/b+c/d=(ad+bc)/bd
    n1=self.num*other.den+other.num*self.den
    n2=self.den*other.den
    return Fraction(n1,n2)
  def sub(self,other):
    #a/b-c/d=(ad-bc)/bd
    n1=self.num*other.den-other.num*self.den
    n2=self.den*other.den
    return Fraction(n1,n2)
  def mul(self,other):
    #a/b*c/d=ac/bd
    n1=self.num*other.num
    n2=self.den*other.den
    return Fraction(n1,n2)
  def div(self,other):
    #(a/b)/(c/d)=ac/bd
    n1=self.num*other.den
    n2=self.den*other.num
    return Fraction(n1,n2)
  def __str__(self):
    return f"{self.num}/{self.den}"
try:
  p=input().split()
  a,b,op,c,d=int(p[0]),int(p[1]),p[2],int(p[3]),int(p[4])
  f1=Fraction(a,b)
  f2=Fraction(c,d)
  result=None
  if op=='+':
    result=f1.add(f2)
  elif op=='-':
    result=f1.sub(f2)
  elif op=='*':
    result=f1.mul(f2)
  elif op=='/':
    result=f1.div(f2)
  if len(p)==5:
    if result is not None:
      print(result)
    else:
      print("Phép tính không hợp lệ!")
  else:
    print("Nhập sai định dạng")
except ValueError as e:
  print(f"Lỗi dữ liệu: {e}")
except ZeroDivisionError:
  print("Lỗi chia cho 0")
```

Lớp BankAccount(owner, balance) với:

deposit(x), withdraw(x); nếu rút quá số dư thì bỏ qua giao dịch.

__str__() in owner: balance.

Input

Dòng 1: owner initial_balance

Dòng 2: n (số lệnh)

Tiếp theo n dòng, mỗi dòng: DEPOSIT x hoặc
WITHDRAW x.

Output
Số dư cuối cùng (in số thực, 2 chữ số)

```python
class BankAccount:
  def __init__(self,owner,balance):
    self.owner=owner
    self.balance=float(balance)
  def deposit(self,amount):
    self.balance+=amount
  def withdraw(self,amount):
    if amount<=self.balance:
      self.balance-=amount
  def __str__(self):
    return f"{self.balance}"
s=input().split()
initial_balance=float(s[-1])
owner=s[0]
myaccount=BankAccount(owner,initial_balance)
n=int(input())
for _ in range(n):
  com=input().split()
  cm=com[0]
  amount=float(com[1])
  if cm=="DEPOSIT":
    myaccount.deposit(amount)
  elif cm=="WITHDRAW":
    myaccount.withdraw(amount)
print(f"{myaccount.balance:.2f}")
```

Yêu cầu

Lớp Student(name) với:

add_score(subject, score)

avg() trả về điểm trung bình.

rank() trả về "Excellent" nếu avg ≥ 8, "Good" nếu 6.5–<8, "Average" nếu 5–<6.5, ngược lại "Poor".

Input
Dòng 1: name

Dòng 2: n

n dòng tiếp: subject score (score là số thực 0–
10)

Output
name avg rank (avg 2 chữ số).


```python
class Student:
  def __init__(self,name):
    self.name=name
    self.scores={}
  def add_score(self,subject,score):
    self.scores[subject]=score
  def avg(self):
    if not self.scores:
      return 0
    return sum(self.scores.values())/len(self.scores)
  def rank(self):
    avg=self.avg()
    if avg>=8:
      return "Excellent"
    elif 6.5<=avg<8:
      return "Good"
    elif 5<=avg<6.5:
      return "Average"
    else:
      return "Poor"
name=input()
n=int(input())
student=Student(name)
for _ in range(n):
  data=input().split()
  subject=data[0]
  score=float(data[1])
  student.add_score(subject,score)
print(f"{name} {student.avg():.2f} {student.rank()}")

```

Lớp Book(title, author, year). Lớp Library() với:

add(book)

count_by_author(name)

find_by_year(y) trả về số sách xuất bản năm y.

Input

Dòng 1:

n (số thao tác)

n dòng:

ADD title;author;year

COUNT author (đếm theo tác giả)

COUNTYEAR year (đếm theo năm)

Output

Mỗi lệnh COUNT/COUNTYEAR in một dòng kết
quả.


```python
class Book:
  def __init__(self,title,author,year):
    self.title=title
    self.author=author
    self.year=int(year)
class Library:
  def __init__(self):
    self.books=[]
  def add(self,book):
    self.books.append(book)
  def count_author(self,name):
    count=0
    for book in self.books:
      if book.author==name:
        count+=1
    return count
  def find_year(self,y):
    count=0
    for book in self.books:
      if book.year==y:
        count+=1
    return count
n=int(input())
lib=Library()
for _ in range(n):
  line=input().strip()
  p=line.split(' ',1)
  cm=p[0]
  if cm=='ADD':
    data=p[1]
    info=data.split(';')
    title=info[0]
    author=info[1]
    year=info[2]
    new_book=Book(title,author,year)
    lib.add(new_book)
  elif cm=='COUNT':
    if len(p)>1:
      author_name=p[1]
      print(lib.count_author(author_name))
  elif cm=='COUNTYEAR':
    if len(p)>1:
      year=int(p[1])
      print(lib.find_year(year))
```

(BCD): n2=BCxCD (Pháp tuyến)
tính góc n1,n2: cos(phi)=(n1xn2)/|n1|x|n2|

```python

import math
class Point:
  def __init__(self,x,y,z):
    self.x=x
    self.y=y
    self.z=z
  def __sub__(self,other):
    #Phép trừ hai điểm tạo thành vector B-A
    return Point(self.x-other.x,self.y-other.y,self.z-other.z)
  def cross(self,other):
    #Tính tích có hướng
    return Point(
        self.y*other.z-self.z*other.y,
        self.z*other.x-self.x*other.z,
        self.x*other.y-self.y*other.x
    )
  def dot(self,other):
    #Tính tích vô hướng
    return self.x*other.x+self.y*other.y+self.z*other.z
  def absolute(self):
    return math.sqrt(self.x**2+self.y**2+self.z**2)
point=[]
for _ in range(4):
  s=input().split()
  diem=[float(val) for val in s]
  point.append(Point(diem[0],diem[1],diem[2]))
a,b,c,d=point[0],point[1],point[2],point[3]
ab=b-a
bc=c-b
cd=d-c
#Tìm vector pháp tuyến của 2 mặt phẳng
#Vector pháp tuyến n1 của (ABC)
x=ab.cross(bc)
#Vector pháp tuyến n2 của (BCD)
y=bc.cross(cd)
#Tính cos(phi)=(x.y)/(|x|*|y|)
num=x.dot(y)
den=x.absolute()*y.absolute()
if den==0:
  print("0.00")
cos_phi=num/den
cos_phi=max(min(cos_phi,1.0),-1.0)
phi_rad=math.acos(cos_phi) #Chuyển về radian
phi_deg=math.degrees(phi_rad) #Chuyển về độ
print(f"{phi_deg:.2f}")
```

Khai báo kiểu dữ liệu biểu diễn số phức (ComplexNumber\) bao gồm \(2\) biến \(realPart\)
và \(imaginaryPart\) kiểu \(double\) biểu diễn phần thực và phần ảo của một số phức.

Viết hàm khởi tạo ComplexNumber(double
initRealPart, double initImaginaryPart) để khởi tạo các giá trị phần thực và phần ảo của một số phức và hàm ComplexNumber multiply(complexNumber
other) trả về tích của số phức hiện tại và một số phức khác.

Đầu vào: Đầu vào từ bàn phím gồm hai dòng, mỗi dòng gồm hai số thực cách nhau bởi một dấu cách biểu diễn phần thực và phần ảo của một số phức.

Các số trong đầu vào có giá trị tuyệt đối không vượt quá \(100\).

Đầu ra
In ra màn hình hai số biểu diễn phần thực và phần ảo của tích hai số phức.
*Phép nhân hai số phức
(a+bi)x(c+di)=(ac-bd)+(ad+bc)i



```python
class ComplexNumber:
  def __init__(self,realPart,imaginaryPart):
    self.realPart=realPart
    self.imaginaryPart=imaginaryPart
  def multiply(self,other):
    real=self.realPart*other.realPart-self.imaginaryPart*other.imaginaryPart
    imaginary=self.realPart*other.imaginaryPart+self.imaginaryPart*other.realPart
    return ComplexNumber(real,imaginary)
a,b=map(int,input().split())
c,d=map(int,input().split())
i1=ComplexNumber(a,b)
i2=ComplexNumber(c,d)
result=i1.multiply(i2)
print(f"{result.realPart} {result.imaginaryPart}")
```

[Triangle]

Khai báo cấu trúc dữ liệu tam giác \(Triangle\)
chứa \(3\) biến \(egde1,\,edge2,\,edge3\) kiểu \
(int\) là độ dài \(3\) cạnh của tam giác đó.

Viết hàm khởi tạo Triangle(int initEdge1,int
initEdge2, int initEdge3) để khởi tạo độ dài các
cạnh của tam giác và hàm double getArea() trả về
diện tích của tam giác đó (nếu tam giác không
hợp lệ, trả về \(-1\)).

Hoàn thiện chương trình dưới đây nhập vào \(3\)
cạnh của một tam giác và in ra diện tích của tam
giác đó.

Đầu vào
Một dòng duy nhất từ bàn phím chứa \(3\) số
nguyên là độ dài \(3\) cạnh của tam giác

Đầu ra
In ra màn hình diện tích tam giác (làm tròn đến\
(2\) chữ số thập phân sau dấu phẩy). In ra \
(invalid\) nếu \(3\) số nhập vào không phải là \(3\)
cạnh của một tam giác


```python
class Triangle:
  def __init__(self,edge1,edge2,edge3):
    self.edge1=edge1
    self.edge2=edge2
    self.edge3=edge3
  def getarea(self):
    if self.edge1+self.edge2>self.edge3 and self.edge1+self.edge3>self.edge2 and self.edge3+self.edge2>self.edge1:
      return True
    else:
      return False
  def dientich(self):
    p=(self.edge1+self.edge2+self.edge3)/2
    return (p*(p-self.edge1)*(p-self.edge2)*(p-self.edge3))**0.5
edge=input().split()
a,b,c=edge[0],edge[1],edge[2]
tamgiac=Triangle(int(a),int(b),int(c))
if tamgiac.getarea():
  print(f"{tamgiac.dientich():.2f}")
else:
  print("invalid")
```
